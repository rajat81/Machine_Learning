```python
X_new = file.iloc[:,2:4]
```

```python
X_new.head()
```

```python
kmeans = KMeans(n_clusters=3, random_state=0).fit(X_new)
```

```python
kmeans.labels_
```

```python
kmeans.cluster_centers_
```

```python
output_kmeans = kmeans.predict(X_new)
plt.scatter(X_new.Age, X_new.EstimatedSalary, c=output_kmeans)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], c='red')
```
